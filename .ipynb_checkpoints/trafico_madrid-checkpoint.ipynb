```python
import xml.etree.ElementTree as ET  
import re
import urllib3
import pandas as pd
import certifi
from datetime import datetime
import smtplib
import numpy as np
import pandas as pd
```

```python
path='http://informo.munimadrid.es/informo/tmadrid/pm.xml'
```

```python
panda_result=pd.DataFrame()
http = urllib3.PoolManager(cert_reqs='CERT_REQUIRED',ca_certs=certifi.where())
    #http = urllib3.PoolManager()
response = http.request('GET', path)
root=ET.fromstring(response.data)
#fecha=root.find('.prefix:updated',ns).text
```

```python
hora=root.find('fecha_hora').text
```

```python
hora
```

```python
for i in root.findall('pm'):
    try:
        idelem=i.find('idelem').text
    except:
        idelem=0
    try:
        descripcion=i.find('descripcion').text
    except:
        descripcion='N/A'
    try:
        intensidad=i.find('carga').text
    except:
        intensidad='N/A'
    print ('Trafico en elemento {},{}, equivalente a {}'.format(idelem, descripcion, intensidad))
```

```python
lista=[]
for i in root.findall('pm'):
    if i.find('error').text=='N':
        entrada={}
        entrada['idelem']=i.find('idelem').text
        try:
            entrada['descripcion']=i.find('descripcion').text
        except:
            entrada['descripcion']='N/A'
        try:
            entrada['intensidad']=i.find('intensidad').text
        except:
            entrada['intensidad']='N/A'
        try:
            entrada['carga']=i.find('carga').text
        except:
            entrada['carga']='N/A'
        try:
            entrada['ocupacion']=i.find('ocupacion').text
        except:
            entrada['ocupacion']='N/A'
        entrada['hora']=hora
        lista.append(entrada)
```

```python
df=pd.DataFrame(lista,columns=['idelem','descripcion','intensidad','carga','ocupacion','hora'])
```

```python
df.set_index(['idelem'],inplace=True)
```

```python
df['intensidad']=df['intensidad'].astype(int)
df['carga']=df['carga'].astype(int)
df['ocupacion']=df['ocupacion'].astype(int)
df['hora']=pd.to_datetime(df['hora'])
```
