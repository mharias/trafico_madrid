```python
import xml.etree.ElementTree as ET  
import re
import urllib3
import pandas as pd
import certifi
from datetime import datetime
import smtplib
```

```python
def xml2pandas(root,CPVs,estado_licitacion,ns):
    
    lista=[]
    
    for a in root.findall('./prefix:entry',ns):
        entrada={}
        
        try:
            status=a.find('.//cbc-place-ext:ContractFolderStatusCode',ns).text
        except:
            status='N/A'
        
        try:
            capitulo=a.find('.//cbc:ItemClassificationCode',ns).text
        except:
            capitulo='00'
        
        try:
            title=a.find('./prefix:title',ns).text
        except:
            title='N/A'
        
        try:
            link=a.find('./prefix:link',ns).attrib['href']
        except:
            link='N/A'
        
        try:
            importe=a.find('.//cbc:TaxExclusiveAmount',ns).text
        except:
            importe='0'
        try:
            entidad=a.find('.//cac:PartyName/cbc:Name',ns).text
        except:
            entidad='n/A'
        
        try:
            poblacion=a.find('.//cbc:CityName',ns).text
        except:
            poblacion='N/A'
        
        try:
            provincia=a.find('.//cbc:CountrySubentity',ns).text
        except:
            provincia='N/A'
            
        try:
            fecha=a.find('.//cbc:EndDate',ns).text
        except:
            fecha='N/A'
        
            
    
        if status==estado_licitacion and capitulo[0:2] in CPVs:
            entrada['status']=status
            entrada['capitulo']=capitulo
            entrada['fecha_entrega']=fecha
            entrada['provincia']=provincia
            entrada['poblacion']=poblacion
            entrada['entidad']=entidad
            entrada['link']=link
            entrada['title']=title
            entrada['status']=status
            entrada['capitulo']=capitulo
            entrada['importe']=importe
            lista.append(entrada)
    
    panda=pd.DataFrame(lista,columns=['title','status','link','importe','entidad','poblacion',
                                      'provincia','fecha_entrega','capitulo'])
    try:
        panda['fecha_entrega']=pd.to_datetime(panda['fecha_entrega'],format='%Y-%m-%d')
    except:
        panda['fecha_entrega']='N/A'
    panda['importe']=pd.to_numeric(panda['importe'])
    return panda



def recurrent_url2pandas(path,CPV,days,ns):
    print ('Ok')
    panda_result=pd.DataFrame()
    http = urllib3.PoolManager(cert_reqs='CERT_REQUIRED',ca_certs=certifi.where())
    #http = urllib3.PoolManager()
    response = http.request('GET', path)
    root=ET.fromstring(response.data)
    fecha=root.find('.prefix:updated',ns).text
    fecha_actualizacion=fecha
    nuevo_path=path
    #print (fecha_actualizacion,'\n',path)

    while fecha==fecha_actualizacion:
        panda_=xml2pandas(root,CPV,'ADJ',ns)
        panda_result = pd.concat([panda_result, panda_], ignore_index=True, sort=False)
        #print (fecha,'\n',nuevo_path)
        print('Tratando {} de fecha {}'.format(nuevo_path,datetime.strptime(fecha[0:10],'%Y-%m-%d').strftime('%d/%m')))
        nuevo_path=root.find('.prefix:link/[@rel="next"]',ns).attrib['href']
        response = http.request('GET', nuevo_path)
        root=ET.fromstring(response.data)
        fecha=root.find('.prefix:updated',ns).text

    return panda_result
    
def panda2text(panda):
    #LF="\n".encode('utf-8')
    texto='From: From Delegaciones'+'\n'
    texto+='To: Delegaciones'+'\n'
    texto+='Subject: Resumen de Licitaciones'+'\n'
    texto+='\n'+'\n'+'Egun on, bo día, buenos días:'+'\n'
    texto+='Resumen de licitaciones publicadas hoy en el portal: '
    texto+='https://contrataciondelestado.es'
    texto+='\n'
    texto+='{} licitaciones por un importe total de {:,.0f}€:'.format(licitaciones.shape[0],licitaciones['importe'].sum())+'\n'
    #print (texto)
    for i in panda.index:
        texto+='\n'
        texto+='----------------------------------------------------------------------------------------------'
        texto+='\n'+'\n'
        texto+=panda.loc[i]['title']+'\n'
        texto+='Importe: {:,.1f}€'.format(panda.loc[i]['importe']).replace(",", "X").replace(".", ",").replace("X", ".")+'\n'
        texto+='Fecha entrega: '+panda.loc[i]['fecha_entrega'].strftime('%d/%m/%Y')+'\n'
        #texto+='Fecha entrega: '+str(licitaciones.loc[i]['fecha_entrega'])+'\n'
        texto+='Entidad: '+panda.loc[i]['entidad']+'\n'
        texto+='Poblacion: '+panda.loc[i]['poblacion']+' / '+panda.loc[i]['provincia']+'\n'
        texto+='\n'
        texto+=panda.loc[i]['link']+'\n'
    #texto=texto.encode('utf-8')
    return texto


def server_connect(account, password, server, port=587):
    if int(port) == 465:    # gmail server
        email_server = smtplib.SMTP_SSL(server, str(port))
    else:
        email_server = smtplib.SMTP(server, port)
        email_server.ehlo()
        email_server.starttls()
    email_server.login(account, password)
    return email_server

```
